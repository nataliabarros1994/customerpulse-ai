```python
# ==================== IMPORTANDO BIBLIOTECAS ====================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

print("✅ Todas as bibliotecas importadas com sucesso!")
```

```python
# ==================== CRIANDO A BASE DE DADOS NOVAMENTE ====================
np.random.seed(42)  # para todo mundo ter o mesmo resultado

clientes = 1000

df = pd.DataFrame({
    "idade": np.random.randint(18, 70, clientes),
    "salario": np.random.randint(1500, 15000, clientes),
    "tempo_cliente": np.random.randint(1, 120, clientes),
    "compras": np.random.randint(0, 50, clientes),
    "cancelou": np.random.choice([0, 1], clientes, p=[0.6, 0.4])  # 40% cancelaram
})

print("Base de dados criada!")
print(df.head())
print("\nQuantidade de clientes que cancelaram:", df["cancelou"].sum())
```

```python
# ==================== SEPARANDO VARIÁVEIS ====================
X = df.drop("cancelou", axis=1)   # todas as colunas menos cancelou
y = df["cancelou"]                # só a coluna que queremos prever

print("✅ X (features) tem", X.shape[1], "colunas")
print("✅ y (target) tem", y.nunique(), "classes (0 = ficou, 1 = cancelou)")
```

```python
# ==================== DIVIDINDO OS DADOS ====================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3,      # 30% para teste
    random_state=42
)

print("Treino:", X_train.shape)
print("Teste:", X_test.shape)
```

```python
# ==================== TREINANDO O MODELO ====================
modelo = RandomForestClassifier(
    n_estimators=100,   # quantidade de árvores
    random_state=42
)

modelo.fit(X_train, y_train)

print("✅ Modelo treinado com sucesso!")
```

```python
# ==================== FAZENDO PREVISÕES ====================
y_pred = modelo.predict(X_test)

# Métricas
acuracia = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)

print(f"🎯 Acurácia do modelo: {acuracia:.1%}")
print(f"📊 AUC: {auc:.3f}")
print("\nRelatório completo:")
print(classification_report(y_test, y_pred))
```

```python
# ==================== MATRIZ DE CONFUSÃO ====================
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_test, y_pred), 
            annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusão')
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.show()
```

```python
# ==================== IMPORTÂNCIA DAS VARIÁVEIS ====================
importancia = pd.DataFrame({
    'Variável': X.columns,
    'Importância': modelo.feature_importances_
}).sort_values('Importância', ascending=False)

print(importancia)

# Gráfico
plt.figure(figsize=(8,5))
sns.barplot(data=importancia, x='Importância', y='Variável')
plt.title('Quais variáveis mais influenciam no cancelamento?')
plt.show()
```

```python
print("🎯 CONCLUSÕES DO PROJETO")
print(f"""
✅ O modelo atingiu {acuracia:.1%} de acurácia
✅ AUC: {auc:.3f}

🔍 Principais insights:
- Salário é a variável mais importante para prever cancelamento
- Tempo como cliente também tem forte influência
- Idade e quantidade de compras têm menos impacto

💡 Recomendações para a empresa:
- Oferecer descontos ou benefícios para clientes com salário mais baixo
- Focar retenção nos primeiros 30-60 dias de cliente
- Criar campanhas personalizadas por faixa salarial
""")
```

```python
import streamlit as st
import pandas as pd
import joblib

st.title("🛡️ CustomerPulse AI - Previsor de Churn")

st.write("Preveja se um cliente vai cancelar o serviço")

# Carregar modelo
modelo = joblib.load('modelo_churn_randomforest.pkl')

# Inputs
idade = st.slider("Idade", 18, 70, 35)
salario = st.number_input("Salário (R$)", 1500, 15000, 4500)
tempo = st.slider("Tempo como cliente (meses)", 1, 120, 12)
compras = st.slider("Compras nos últimos 3 meses", 0, 50, 8)

if st.button("🔮 Prever Cancelamento"):
    dados = pd.DataFrame({
        "idade": [idade],
        "salario": [salario],
        "tempo_cliente": [tempo],
        "compras": [compras]
    })
    
    pred = modelo.predict(dados)[0]
    prob = modelo.predict_proba(dados)[0][1]
    
    if pred == 1:
        st.error(f"🚨 ALTO RISCO DE CANCELAMENTO ({prob:.1%})")
    else:
        st.success(f"✅ Baixo risco de cancelamento ({prob:.1%})")

        
```
